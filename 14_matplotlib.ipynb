### Matplotlib

Matplotlib é uma biblioteca de plotagem. Nesta aula, faremos uma breve introdução ao módulo **`matplotlib.pyplot`**, que fornece um sistema de plotagem semelhante ao do MATLAB.


### Plotting

A função mais importante em matplotlib e **`plot`**, que permite plotar dados em 2D. A seguir, um exemplo bem simples:

```
import numpy as np
import matplotlib.pyplot as plt

# Calcular as coordenadas de x e y para pontos em uma curva senóide.
x = np.arange(0, 3 * np.pi, 0.01)
y = np.sin(x)

# Plotar os pontos usando matplotlib
plt.plot(x, y)
plt.show()  # Você deve chamar plt.show() para fazer o gráfico aparecer.
```

Com apenas um pouco de trabalho extra, podemos plotar facilmente várias linhas ao mesmo tempo e adicionar um título, legenda e rótulos de eixo:

```
import numpy as np
import matplotlib.pyplot as plt

# Calcule as coordenadas x e y para os pontos das curvas senóide e cossenóide.
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plote os pontos usando matplotlib.
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('rótulo do eixo x')
plt.ylabel('rótulo do eixo y')
plt.title('Seno e Cosseno')
plt.legend(['Seno', 'Cosseno'])
plt.show()
```

### Subplots

Você pode plotar coisas diferentes na mesma figura usando a função **`subplot`**. Aqui está um exemplo:

```
import numpy as np
import matplotlib.pyplot as plt

# Calcular as coordenadas x e y para os pontos das curvas senóide e cossenóide.
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Configure uma grade de subplot que tenha altura 2 e largura 1,
# e defina o primeiro subplot como ativo.
plt.subplot(2, 1, 1)

# Faça a primeira plotagem
plt.plot(x, y_sin)
plt.title('Seno')

# Configure a segunda subplot como ativa, e faça a segunda plotagem.
plt.subplot(2, 1, 2)
plt.plot(x, y_cos)
plt.title('Cosseno')

# Mostre a figura construída.
plt.show()
```
